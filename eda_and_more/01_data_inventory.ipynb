# 01 Data Inventory

This notebook inventories the raw seed files and summarizes the field structure used by the graph model.

## Analysis Focus

- entity counts by file
- field coverage by entity type
- presence of `tags`
- presence of `...Refs` fields

```python
import json
from pathlib import Path

seed_dir = Path('../seed-data')
files = sorted(seed_dir.glob('*.json'))
[(file.name, len(json.load(open(file)))) for file in files]
```

```python
for file in files:
    data = json.load(open(file))
    print(file.name)
    print(sorted(data[0].keys()))
    print()
```
